```python
# EDA for startup data.csv

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

sns.set(style="whitegrid", context="notebook")

raw_path = Path('startup data.csv')
if not raw_path.exists():
    raise FileNotFoundError("Expected 'startup data.csv' in project root.")

df = pd.read_csv(raw_path, low_memory=False)
print(f"Raw shape: {df.shape}")

df.columns = [c.strip() for c in df.columns]
```

```python
# Overview

display(df.head())
display(df.sample(5, random_state=42))
display(df.describe(include='all').transpose().head(10))
```

```python
# Missingness overview

missing = df.isna().mean().sort_values(ascending=False)
print("Top 20 columns by missingness:")
display(missing.head(20))
plt.figure(figsize=(8,4))
missing.head(20).plot(kind='barh')
plt.title('Top 20 Missingness')
plt.tight_layout();
```

```python
# Status distribution

if 'status' in df.columns:
    plt.figure(figsize=(4,3))
    sns.countplot(x=df['status'].astype(str).str.lower())
    plt.title('Status distribution')
    plt.tight_layout();
```

```python
# Numeric correlations

num_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
if len(num_cols) > 1:
    corr = df[num_cols].corr()
    plt.figure(figsize=(10,8))
    sns.heatmap(corr, cmap='vlag', center=0, linewidths=0.1)
    plt.title('Numeric correlation heatmap')
    plt.tight_layout();
```

```python
# Numeric distributions

key_numeric = [c for c in ['funding_total_usd','funding_rounds','milestones','avg_participants','age_first_funding_year','age_last_funding_year'] if c in df.columns]
for c in key_numeric:
    plt.figure(figsize=(5,3))
    sns.histplot(df[c], bins=40, kde=True)
    plt.title(f'Distribution of {c}')
    plt.tight_layout();
```

```python
# Categorical summaries

cat_cols = [c for c in ['state_code','category_code','city'] if c in df.columns]
for c in cat_cols:
    top = df[c].astype(str).str.lower().value_counts().head(15)
    plt.figure(figsize=(6,3))
    sns.barplot(x=top.values, y=top.index)
    plt.title(f'Top {c} values')
    plt.tight_layout();
```

```python
# Relationships with target (if available)

if 'status' in df.columns:
    target = df['status'].astype(str).str.lower().map({'acquired':1,'operating':0,'closed':0})
    if target.notna().any():
        for c in [c for c in ['funding_total_usd','funding_rounds','milestones','avg_participants'] if c in df.columns]:
            plt.figure(figsize=(5,3))
            sns.boxplot(x=target, y=df[c])
            plt.title(f'{c} vs target (0/1)')
            plt.xlabel('target')
            plt.tight_layout();
print("EDA sections ready.")
```
